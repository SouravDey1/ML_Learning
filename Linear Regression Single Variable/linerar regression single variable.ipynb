```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
import pandas as pd

data = {
    'area': [1000, 1500, 1800, 2400, 3000],
    'price': [300000, 450000, 500000, 600000, 650000]
}

df = pd.DataFrame(data)
df.to_csv("homeprices.csv", index=False)
print("Sample homeprices.csv created!")
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
import pandas as pd
reg.predict(pd.DataFrame({'area': [2150]}))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import matplotlib.pyplot as plt

# For Jupyter Notebook (optional)
%matplotlib inline

plt.xlabel('Area (sqft)', fontsize=20)
plt.ylabel('Price', fontsize=20)

plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
plt.title('Home Price Prediction')
plt.grid(True)
plt.show()
```
